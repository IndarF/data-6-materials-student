### [Tutorial] Indexing and Slicing

Strings can be viewed as special sequences of characters. Just like arrays, they can be **indexed** and **sliced**. We define each of these features below.

````python
# just run this cell
our_string = "Data Science at Berkeley rocks!"
our_string
````

**Indexing** into a string is getting the character at a particular index of the string. Strings are zero-indexed.

Unlike arrays, strings do not have an `item` method. However, we can use **square bracket notation** to index into strings:

````python
our_string[0] + our_string[5]
````

We can also use negative indexing:

````python
our_string[-1]
````

**Slicing** into a string, or getting a **slice** of a string, is getting a sequence of characters between two indexes. Slices also use square brackets, but now we specify the starting and ending indices of the range:

```python
s[start:end]
```

Similar to array ranges (e.g., `np.arange`), the starting index is included in the slice, and the ending index is excluded from the slice.

````python
our_string[16:24]
````

**Slicing, full square bracket notation**:
While slices by default will take every character in the `[start, end)` range of indices, square bracket notation can also specify the `step`, or number of characters to skip. This behavior is just like array ranges.

```python
s[start:end:step]
```

````python
our_string[1:-2:2]
````
