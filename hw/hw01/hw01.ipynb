# Question 4 – Markdown

[Markdown](https://en.wikipedia.org/wiki/Markdown) isn't a programming language – it's a "markup language", which means that it is designed to format text to look nice. But like a programming language, Markdown has a set of rules (syntax) that govern how it works. Think of it like a fancy Google Docs.

Here are some of the most important rules in Markdown:

- Double asterisks (`**`) are how we make words bold. `My name is **Jordan**` renders to "My name is **Jordan**".
- Underscores (`_`) are how we make words italicized. `How are _you_?` renders to How are _you_?, and `what is **_up_**` renders to what is **_up_**.
- Hashtags (`#`, `##`, `###`) are how we create titles/headings. The more hashtags you include, the smaller the title is. The title for this question was generated using `## Question 5 – Markdown`.
- Bulleted lists are created using the (`-`) character.
- Numbered lists are created the same way they're created in a standard word processor, like Google Docs.

Markdown is important to learn because Jupyter Notebooks are often full of Markdown cells (just like this assignment). In future assignments, we will ask you to write answers to written questions using Markdown, and while the majority of your answers will be in plaintext (i.e. with none of the special characters above), learning Markdown is valuable nonetheless.

# Question 7 — Age-Standardized Rates

An **age-standardized rate** levels the field of comparison by restructuring both populations to have the same age structure, often to a **standard population**. The article walks through this computation using a standard population from the 1991 Census to obtain:

> [the] 2000 age-standardized mortality rate of 177.9 deaths per 100,000 standard population, and
> 
> [the] 2011 age-standardized mortality rate of 163.6 deaths per 100,000 standard population.

(It is worth noting that the numbers computed in the article have significant roundoff, but the idea still stands.)

The article footnotes further mention that a "transition to a more recent age structure is being considered." Let's now consider transitioning the age standard to the more recent 2021 census.

For reference, here are the (rounded) age-specific crude rates presented in Table 1 of the article:

| Age Group | 2000 | 2011 |
| --- | --- | --- |
| 0 to 39 years | 7.9 | 5.8 |
| 40 years and over | 450.4 | 416.7 |
| Total all ages | 204.2 | 211.0 |

We have also constructed the table below, which lists the proportions of the populations in each age group for both 1991 and **2021** ([source](https://www12.statcan.gc.ca/census-recensement/2021/dp-pd/prof/details/page.cfm?Lang=E&DGUIDList=2021A000011124&GENDERList=1,2,3&STATISTICList=1,4&HEADERList=0&SearchText=Canada)).

| Age Group | 1991 | 2021 | 
| --- | --- | --- |
| 0 to 39 years | 61.6% | 47.8% |
| 40 years and over | 38.4% | 52.2% |
                                                                           
The article computes the **Age-Standardized Mortality Rate (ASMR)** as the combination of the age-specific rates, multiplied by the proportion of the standard population. 
For a 1991 standard population, these rates are (as per the article):

$$ ASMR_{2000} = (7.9 \times 61.6\%) + (450.4 \times 38.4\%) = 177.8 $$
$$ ASMR_{2011} = (5.8 \times 61.6\%) + (416.7 \times 38.4\%) = 163.6, $$

where $ASMR_{2000}$ and $ASMR_{2011}$ are the ASMR rates for 2000 and 2011, respectively, computed using the 1991 standard population proportions.
