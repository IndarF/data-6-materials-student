# Part 3: Body Mass Index

From the [National Institute of Health (NIH)](https://www.nhlbi.nih.gov/calculate-your-bmi): 
> **Body mass index (BMI)** is a measure of body fat based on height and weight that applies to adult men and women. Your BMI is just one piece of the puzzle. It‚Äôs based on height and weight but doesn‚Äôt take into account your muscle mass, bone density, or body composition. Your healthcare provider will consider whether your BMI is too high or too low for you.

The last quoted sentence packs a lot of information and is the focus of this section, where we analyze **BMI as a measure of health**.
**BMI** is defined according to the following formula (where 703 comes from converting to the correct units, kilograms per square meter):

$$ \text{BMI} \approx \frac{\text{mass in lb}}{(\text{height in inches})^2} \times 703 $$

and **BMI categories** (as per the [World Health Organization](https://www.who.int/data/gho/data/themes/topics/topic-details/GHO/body-mass-index)) are different body fat classes based on BMI ranges. For adults over 20:

| BMI Range        | BMI Category        |
|------------------|---------------------|
| Less than 18.5   | Underweight         |
| 18.5 to <25      | Healthy             |
| 25 to <30        | Overweight          |
| 30 or greater    | Obesity             |

These BMI categories are used widely in United States healthcare systems to predict a patient's obesity. Obesity is associated with health risks such as diabetes; therefore, it is useful to have a predictor for a person's obesity for subsequent treatment and eligibility for different medical procedures. However, we shall see that BMI (and its categories) falls short when it comes to **measurement reliability and validity**.
