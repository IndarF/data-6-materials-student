## Keep track of counters and totals outside of the loop

````python
# what does this code do?
def average(arr):
    total = 0
    count = 0
    for element in arr:
        total += element       # shorthand for    total = total + element
        count += 1             # shorthand for    count = count + 1
    print("total", total)
    print("count", count)
    return total/count
````

````python
test1 = make_array(1, 2, 3, 4)
average(test1)
````

````python
np.average(test1)
````

````python
average(np.arange(10))
````

# While-loops

## Repeat while the condition is *true*

````python
counter = 10
while counter > 0:
    print(counter)
    counter -= 1
print("blast off!")
````

### Discussion: What is the difference between these two loops?

Every for-loop can be written as a while-loop, but not vice versa!

````python
def sum_squares_for(values):
    total = 0
    for v in values:
        total += v ** 2
    return total

sum_squares_for(make_array(3, 4, 5))
````

````python
def sum_squares_while(values):
    total = 0
    i = 0
    while i < len(values):
        total += values.item(i) ** 2
        i += 1
    return total

sum_squares_while(make_array(3, 4, 5))
````

## Practice: More loops

Find the historic year where UC Berkeley enrolled a cumulative number of 1,000,000 undergraduates.

Source: https://www.universityofcalifornia.edu/about-us/information-center/historical-enrollment

````python
enrollment = Table.read_table("data/ucberkeley_enrollment.csv")
enrollment
````

````python
total_undergrads = 0
year_index = 0
undergrads = enrollment.column("undergrad")

while total_undergrads <= 1000000:
    total_undergrads += undergrads.item(year_index)
    year_index += 1

print('Total undergrads graduated by', enrollment.column("year").item(year_index),
      "was", total_undergrads)
````
